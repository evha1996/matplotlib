```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib.patches as pch
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# city_data.head()
# ride_data.head()

# Combine the data into a single dataset
combined = pd.merge(ride_data, city_data, on="city")


combined
# Display the data table for preview
```

```python
new = combined.dropna() #getting rid of null values 


new
```

## Bubble Plot of Ride Sharing Data

```python
city_type = new[['city', 'type']]
city_type = city_type.groupby('type').count().reset_index()


city_type 
```

```python
cities = new[['city','ride_id']]

cities_rides = cities.groupby('city').nunique()


cities

cities_fares = new[['city', 'fare']]

cities_fares = cities_fares.groupby('city').mean()

cities_fares

cities_drivers = new[['city','driver_count']]

cities_drivers = cities_drivers.groupby('city').sum()

cities_drivers


```

```python
colors = city_type.type.map({'Urban':'cyan', 'Suburban':'magenta', 'Rural':'yellow'})
plt.scatter(x = cities_rides['ride_id'], y = cities_fares['fare'], s = cities_drivers['driver_count'] , color = colors, alpha = 0.7, edgecolor='black', linewidths=1.5)
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")
plt.ylim([15, 45])
plt.yticks(np.arange(15,45,5))
types = ['Urban','Suburban','Rural']
circle_1 = pch.Circle((0.7, 0.7), 1, facecolor='cyan', linewidth=3)
circle_2 = pch.Circle((0.7, 0.7), 1, facecolor='magenta', linewidth=3)
circle_3 = pch.Circle((0.7, 0.7), 1, facecolor='yellow', linewidth=3)
lgnd = plt.legend((circle_1,circle_2,circle_3),("Urban", "Suburban", "Rural"), numpoints=1, loc='upper right', title='City Types')

plt.grid()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percent

city_shares = new.groupby('type').sum()

city_shares

city_shares['fare'].sum()


city_shares['city_share'] = city_shares['fare']/63538.63

city_shares

# city_shares

#63538.63999999999
# Build Pie Chart

# Save Figure

```

```python
# Show Figure

labels = 'Rural', 'Suburban', 'Urban'
explode = (0, 0, 0.1) 

colors = ['yellow', 'magenta', 'cyan']

plt.pie(city_shares['city_share'], explode=explode, labels=labels, colors = colors , autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.show()
```
